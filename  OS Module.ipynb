# Chapter 1: Introduction to the OS Module

## What is the OS Module?

### Definition
The **OS (Operating System) module** is a built-in Python module that provides functions to interact with the operating system. It allows Python programs to create, delete, rename, move, and manage files and folders, as well as perform many system-related tasks.

---

### Simple Meaning

The **OS module acts as a bridge between Python and your computer's operating system.**

It allows Python to communicate directly with your computer.

---

### One-Line Definition

> **The OS module enables Python programs to perform operating system-related tasks such as managing files, folders, directories, and system information.**

---

## Real-World Analogy

Imagine Python is an employee in a company.

The employee cannot directly open office doors, create rooms, or remove furniture.

Instead, the employee requests the **office manager (Operating System)** to perform those tasks.

Similarly,

- Python → Employee
- Operating System → Office Manager
- OS Module → Communication bridge

Python sends requests through the OS module, and the Operating System performs the task.

---

## Why Do We Use the OS Module?

Without the OS module, Python cannot easily perform system-related operations.

The OS module helps us:

- Create folders
- Delete folders
- Rename files
- Rename directories
- Check the current working directory
- List files and folders
- Create nested folders
- Delete nested folders
- Traverse directory structures
- Check whether a file exists
- Work with file paths
- Execute system commands

# Chapter 4: Current Working Directory

## getcwd()

### Definition

Returns the current working directory.

The current working directory is the folder from which your Python program is running.

```python
os.getcwd()
```

# Chapter 5: Create a Folder

## mkdir()

### Definition

Creates a single new directory (folder).

---

### Syntax

```python
os.mkdir("folder_name")
```

---

### Example

```python
import os

os.mkdir("Python")
```

Creates

```
Python/
```

---

## Conditions

### Condition 1

The parent directory must exist.

Example

```
Projects/
```

Inside it, you can create

```
Python/
```

---

### Condition 2

The folder should NOT already exist.

---

## Error Cases

### Parent Doesn't Exist

```python
os.mkdir("Parent/Python")
```

Error

```
FileNotFoundError
```

---

### Folder Already Exists

```python
os.mkdir("Python")
```

Error

```
FileExistsError
```

---

### Safe Method

```python
import os

if not os.path.exists("Python"):
    os.mkdir("Python")
```

# Chapter 9: Create Multiple Folders

## makedirs()

### Definition

Creates multiple nested folders automatically.

---

### Syntax

```python
os.makedirs("Parent/Child/GrandChild")
```

---

### Example

```python
import os

os.makedirs("College/CSE/Python")
```

Creates

```
College
    |
    CSE
       |
       Python
```

---

## Advantage

Unlike `mkdir()`,

`makedirs()` automatically creates missing parent directories.

---

## Conditions

If the complete path already exists,

```
FileExistsError
```

---

## Safe Method

```python
import os

os.makedirs("College/CSE/Python", exist_ok=True)
```

`exist_ok=True` prevents an error if the directory already exists.

---

# Chapter 11: Walk Through Directories

## walk()

### Definition

Traverses every folder and file inside a directory.

Useful for searching, indexing, backups, and automation.

---

### Syntax

```python
os.walk(path)
```

---

### Example

```python
import os

for root, dirs, files in os.walk("."):
    print("Current Folder :", root)
    print("Directories    :", dirs)
    print("Files          :", files)
    print()
```

---

### Example Output

```
Current Folder : .
Directories : ['Images', 'Python']
Files : ['main.py']

Current Folder : ./Python
Directories : []
Files : ['test.py']
```

---

## Conditions

The starting directory must exist.

Otherwise,

```
FileNotFoundError
```

# Chapter 20: Best Practices

Always check before creating a folder.

```python
if not os.path.exists("Python"):
    os.mkdir("Python")
```

---

Always check before deleting.

```python
if os.path.exists("Python"):
    os.rmdir("Python")
```

---

Use `os.path.join()` instead of manually writing slashes.

```python
path = os.path.join("Folder", "File.txt")
```

---

Use `exist_ok=True` with `makedirs()`.

```python
os.makedirs("Python/Notes", exist_ok=True)
```

---

Never delete folders without checking.

# Interview Questions

### 1. What is the OS module?

The OS module is a built-in Python module used to interact with the operating system and perform file, directory, and system-level operations.

---

### 2. Difference between `mkdir()` and `makedirs()`?

- `mkdir()` creates only one folder.
- `makedirs()` creates multiple nested folders automatically.

---

### 3. Difference between `rmdir()` and `removedirs()`?

- `rmdir()` removes one empty folder.
- `removedirs()` removes multiple nested empty folders.

---

### 4. What does `os.getcwd()` return?

It returns the current working directory.

---

### 5. What does `os.listdir()` return?

It returns a list containing all files and folders in the specified directory.

---

### 6. What is the difference between `.` and `..`?

- `.` → Current directory
- `..` → Parent directory

---

### 7. How do you check if a file or folder exists?

Using:

```python
os.path.exists(path)
```

---

### 8. Why should you use `os.path.join()`?

It creates platform-independent file paths and automatically uses the correct path separator for Windows, Linux, or macOS.

# Real-World Teaching Line

Before performing any file or folder operation:

- Check whether it exists.
- Check whether it is empty (if deleting).
- Check whether the destination already exists (if renaming or creating).
- Use safe methods like `os.path.exists()` and `exist_ok=True`.

These checks make your Python programs more reliable and prevent unexpected errors.
