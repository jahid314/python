# 1. Introduction

In software development, data often needs to be:

* Stored in files
* Saved in databases
* Sent over networks
* Shared between applications
* Cached for future use

Python objects cannot directly travel across networks or be stored in files efficiently.

Therefore, they must be converted into a transferable format.

This process is called **Serialization**.

---

# 2. What is Serialization?

## Definition

Serialization is the process of converting a Python object into a format that can be:

* Stored
* Transmitted
* Shared

### Example

```python
user = {
    "name": "Yogi",
    "age": 25
}
```

Serialized form:

```json
{
    "name": "Yogi",
    "age": 25
}
```

---

# 3. What is Deserialization?

## Definition

Deserialization is the process of converting serialized data back into its original Python object.

### Example

JSON:

```json
{
    "name": "Yogi",
    "age": 25
}
```

After Deserialization:

```python
{
    'name': 'Yogi',
    'age': 25
}
```

---

### Mapping

| Real World | Programming     |

# 8. PICKLE MODULE

---

## What is Pickle?

Pickle converts Python objects into binary format.

### Import

```python
import pickle
```

---

## dump()

Write object into file.

```python
import pickle

data = {
    "name": "Yogi",
    "age": 25
}

with open("data.pkl", "wb") as file:
    pickle.dump(data, file)

print("Data Saved")
```

---

## load()

Read object from file.

```python
import pickle

with open("data.pkl", "rb") as file:
    data = pickle.load(file)

print(data)
```

Output:

```python
{'name': 'Yogi', 'age': 25}
```

---

## dumps()

Convert object into bytes.

```python
import pickle

data = {
    "name": "Yogi"
}

binary_data = pickle.dumps(data)

print(binary_data)
```

Output:

```python
b'\x80\x04...'
```

---

## loads()

Convert bytes back into object.

```python
import pickle

binary_data = pickle.dumps(
    {"name": "Yogi"}
)

data = pickle.loads(binary_data)

print(data)
```

Output:

```python
{'name': 'Yogi'}
```

---

## Pickle Functions Summary

| Function | Purpose        |
| -------- | -------------- |
| dump()   | Object → File  |
| load()   | File → Object  |
| dumps()  | Object → Bytes |
| loads()  | Bytes → Object |

---

## Advantages

* Very Fast
* Supports Complex Objects
* Easy to Use
* Efficient Storage

---

## Disadvantages

* Python Only
* Not Human Readable
* Not Secure

# 9. PICKLE SECURITY WARNING

Never load pickle data from unknown sources.

❌ Bad

```python
pickle.load(file_from_internet)
```

Reason:

```text
Malicious code may execute.
```

Only use pickle with trusted data.

# 10. PICKLE PROTOCOLS

Pickle uses protocols internally.

Example:

```python
import pickle

data = {"name": "Yogi"}

with open("data.pkl", "wb") as file:
    pickle.dump(data, file, protocol=5)
```

Higher protocol:

* Faster
* Smaller files
* Better performance

# 11. JSON MODULE

---

## What is JSON?

JSON stands for:

```text
JavaScript Object Notation
```

It is the most popular format for data exchange.

---

## Import

```python
import json
```

---

## dump()

Write object to file.

```python
import json

data = {
    "name": "Yogi",
    "age": 25
}

with open("data.json", "w") as file:
    json.dump(data, file)
```

---

## load()

Read from file.

```python
import json

with open("data.json", "r") as file:
    data = json.load(file)

print(data)
```

---

## dumps()

Convert object to JSON string.

```python
import json

data = {
    "name": "Yogi"
}

json_data = json.dumps(data)

print(json_data)
```

Output:

```json
{"name": "Yogi"}
```

---

## loads()

Convert JSON string into Python object.

```python
import json

json_data = '{"name":"Yogi"}'

data = json.loads(json_data)

print(data)
```

Output:

```python
{'name': 'Yogi'}
```

---

## Pretty Printing JSON

```python
import json

data = {
    "name": "Yogi",
    "age": 25
}

print(
    json.dumps(
        data,
        indent=4
    )
)
```

Output:

```json
{
    "name": "Yogi",
    "age": 25
}
```

# 13. JSON Unsupported Data Types

JSON cannot directly serialize:

```python
set
tuple
complex
datetime
custom objects
```

Example:

```python
import json

data = {1, 2, 3}

json.dumps(data)
```

Output:

```python
TypeError
```

---

# Converting Unsupported Types

Example:

```python
import json

data = {1, 2, 3}

json_data = json.dumps(list(data))

print(json_data)
```

Output:

```json
[1, 2, 3]
```

## What is YAML?

```text
YAML Ain't Markup Language
```

---

## Installation

```bash
pip install pyyaml
```

---

## Import

```python
import yaml
```

---

## dump()

```python
import yaml

data = {
    "name": "Yogi",
    "age": 25
}

with open("data.yaml", "w") as file:
    yaml.dump(data, file)
```

---

## safe_load()

```python
import yaml

with open("data.yaml", "r") as file:
    data = yaml.safe_load(file)

print(data)
```

---

## Memory Example

```python
import yaml

data = {
    "name": "Yogi"
}

yaml_data = yaml.dump(data)

print(yaml_data)

new_data = yaml.safe_load(yaml_data)

print(new_data)
```

---

## Why safe_load()?

Always use:

```python
yaml.safe_load()
```

Avoid:

```python
yaml.load()
```

Reason:

```text
safe_load() prevents unsafe code execution.
```

# 15. dump() vs dumps()

Easy Rule:

```text
dump  = File
dumps = Memory
```

---

## dump()

```python
json.dump(data, file)
pickle.dump(data, file)
yaml.dump(data, file)
```

Writes to file.

---

## dumps()

```python
json.dumps(data)
pickle.dumps(data)
yaml.dump(data)
```

Returns string/bytes in memory.

---

# 16. load() vs loads()

Easy Rule:

```text
load  = File
loads = Memory
```

---

## load()

```python
json.load(file)
pickle.load(file)
yaml.safe_load(file)
```

Reads from file.

---

## loads()

```python
json.loads(json_string)
pickle.loads(binary_data)
yaml.safe_load(yaml_string)
```

Reads from memory.
