# Exception Handling in Python

## What is Exception Handling?

### Definition

An **Exception** is an error that occurs while a program is running (Runtime Error).

**Exception Handling** is a mechanism that allows a program to detect, handle, and recover from runtime errors without terminating unexpectedly.

### Example

```python
a = 10
b = 0

print(a / b)
```

**Output**

```python
ZeroDivisionError: division by zero
```

Without exception handling, the program stops immediately.

---

# Why Do We Need Exception Handling?

In real-world applications such as:

- Banking Systems
- E-Commerce Platforms
- Food Delivery Applications
- Hospital Management Systems
- Social Media Applications

An unexpected error should not crash the entire application.

### Benefits

✅ Prevents application crashes

✅ Improves user experience

✅ Maintains application stability

✅ Protects data integrity

✅ Handles unexpected situations gracefully

✅ Helps in debugging and logging errors

---

# System-Level Understanding

When Python executes a program:

```text
Program Starts
      ↓
Execute Statement
      ↓
Error Occurred?
      ↓
   No → Continue Execution
      ↓
   Yes
      ↓
Create Exception Object
      ↓
Search for Handler
      ↓
Handler Found?
      ↓
Yes → Handle Error
No  → Terminate Program
```

### Example

```python
print(10 / 0)
```

Internal Process:

```text
1. Python detects invalid operation.
2. Creates ZeroDivisionError object.
3. Raises exception.
4. Searches for matching except block.
5. If found → Handle error.
6. If not found → Program terminates.
```

# Components of Exception Handling

## 1. try Block

### What?

Contains code that may generate an exception.

### Why?

Python monitors this block for errors.

### Syntax

```python
try:
    risky_code
```

### Example

```python
try:
    result = 10 / 0
```

---

## 2. except Block

### What?

Handles exceptions generated inside the try block.

### Why?

Prevents application crashes.

### Syntax

```python
except:
    handle_error
```

### Example

```python
try:
    print(10 / 0)

except:
    print("Error occurred")
```

---

## 3. else Block

### What?

Executes only if no exception occurs.

### Why?

Keeps success logic separate from error handling.

### Syntax

```python
try:
    code

except:
    handle_error

else:
    success_code
```

### Example

```python
try:
    num = 10

except:
    print("Error")

else:
    print("Success")
```

---

## 4. finally Block

### What?

Executes regardless of whether an exception occurs.

### Why?

Used for cleanup operations.

### Syntax

```python
finally:
    cleanup_code
```

### Example

```python
try:
    print("Running")

finally:
    print("Cleanup")
```

### Common Uses

- Closing files
- Closing database connections
- Releasing network resources
- Cleaning temporary data

# PART 1: NESTED EXCEPTION HANDLING

# What is Nested Exception Handling?

## Definition

**Nested Exception Handling** means placing one `try-except` block inside another `try-except` block.

In simple words:

```text
try
 └── try
      └── risky code
```

A try block can contain:

- Another try block
- Multiple try blocks
- Multiple except blocks
- finally blocks

---

# Why Do We Need Nested Exception Handling?

In real-world applications, a single task may involve multiple operations.

Example:

```text
Login Process
      ↓
Read User Data
      ↓
Connect Database
      ↓
Validate Password
      ↓
Generate Token
```

Each step can fail for different reasons.

Instead of handling everything in one place, we handle errors at the level where they occur.

---

# Real-Life Example

Suppose you go to an ATM.

```text
Insert Card
      ↓
Enter PIN
      ↓
Check Balance
      ↓
Withdraw Money
```

Possible Errors:

- Card unreadable
- Wrong PIN
- Server unavailable
- Insufficient balance

Each stage can have its own error handling.

This is the idea behind Nested Exception Handling.

---

# Basic Syntax

```python
try:

    try:
        risky_code

    except:
        handle_inner_error

except:
    handle_outer_error
```

---

# Example 1

    try:

## 1. Secure File Manager
What Does It Do?
- Creates files
- Deletes files
- Renames files
- Lists files
- Handles errors safely
## Why Use It?

Similar to:

- Windows File Explorer
- Google Drive Backend
- Dropbox

````python
import os

while True:
    print("\n===== FILE MANAGER =====")
    print("1. Create File")
    print("2. Delete File")
    print("3. Rename File")
    print("4. View Files")
    print("5. Exit")

    choice = input("Enter Choice: ")

    try:

        if choice == "1":
            filename = input("Enter File Name: ")

            with open(filename, "w") as file:
                file.write("File Created Successfully")

            print("File Created")

        elif choice == "2":
            filename = input("Enter File Name: ")

            os.remove(filename)
            print("File Deleted")

        elif choice == "3":
            old_name = input("Old File Name: ")
            new_name = input("New File Name: ")

            os.rename(old_name, new_name)

            print("File Renamed")

        elif choice == "4":
            files = os.listdir()

            for file in files:
                print(file)

        elif choice == "5":
            break

        else:
            print("Invalid Choice")

    except FileNotFoundError:
        print("File Not Found")

    except PermissionError:
        print("Permission Denied")

    except Exception as e:
        print("Error:", e)
````

# 2. Employee Record System
Company Scenario

HR department stores employee data.

Features:

- Add Employee
- View Employees
- Search Employee

````python
import os

FILE_NAME = "employees.txt"

while True:

    print("\n1.Add Employee")
    print("2.View Employees")
    print("3.Search Employee")
    print("4.Exit")

    choice = input("Enter Choice: ")

    try:

        if choice == "1":

            emp_id = input("Employee ID: ")
            name = input("Name: ")

            with open(FILE_NAME, "a") as file:
                file.write(f"{emp_id},{name}\n")

            print("Employee Added")

        elif choice == "2":

            with open(FILE_NAME, "r") as file:

                for data in file:
                    print(data.strip())

        elif choice == "3":

            search_id = input("Enter Employee ID: ")

            found = False

            with open(FILE_NAME, "r") as file:

                for data in file:

                    if data.startswith(search_id):

                        print(data.strip())
                        found = True

            if not found:
                print("Employee Not Found")

        elif choice == "4":
            break

    except FileNotFoundError:
        print("Employee File Missing")

    except Exception as e:
        print(e)
````

# 3. Log Cleaner System
Real Company Usage

Every company generates:

- logs
- error logs
- access logs
- security logs

Old logs consume storage.

This program automatically deletes old log files.

````python
import os

try:

    files = os.listdir()

    for file in files:

        if file.endswith(".log"):

            os.remove(file)

            print(file, "Deleted")

except FileNotFoundError:
    print("File Missing")

except PermissionError:
    print("Permission Denied")

except Exception as e:
    print(e)
````

# 5. Student Result Management System

- OS Module

````python
import os

FILE = "results.txt"

while True:

    print("\n1.Add Result")
    print("2.View Results")
    print("3.Exit")

    choice = input("Choice: ")

    try:

        if choice == "1":

            name = input("Student Name: ")
            marks = int(input("Marks: "))

            if marks < 0 or marks > 100:
                raise ValueError(
                    "Marks must be between 0 and 100"
                )

            with open(FILE, "a") as file:
                file.write(f"{name}-{marks}\n")

        elif choice == "2":

            with open(FILE, "r") as file:

                print(file.read())

        elif choice == "3":
            break

    except ValueError as e:
        print(e)

    except FileNotFoundError:
        print("Result File Missing")
````

# 7. ATM System (Custom Exception + Nested Try)
Concepts Used
- Nested Try
- Custom Exception
- Business Logic

````python
class InsufficientBalance(Exception):
    pass

balance = 5000

try:

    try:

        amount = int(input("Withdraw Amount: "))

        if amount > balance:

            raise InsufficientBalance(
                "Insufficient Balance"
            )

        balance -= amount

        print("Withdrawal Successful")

    except ValueError:

        print("Enter Valid Number")

except InsufficientBalance as e:

    print(e)

finally:

    print("Thank You For Using ATM")
````

# 8. Project Cleanup Tool
Real DevOps Usage

Removes:

- temp files
- cache files
- unwanted files

````python
import os

try:

    files = os.listdir()

    for file in files:

        if file.endswith(".tmp"):

            os.remove(file)

            print(file, "Removed")

except Exception as e:

    print(e)
````

# 10. Mini Bank Management System

This combines:

✅ Exception Handling

✅ Nested Try

✅ Custom Exceptions

✅ File Handling

✅ OS Module

````python
import os

class InsufficientBalance(Exception):
    pass

balance = 10000

while True:

    print("\n1.Check Balance")
    print("2.Withdraw")
    print("3.Exit")

    choice = input("Enter Choice: ")

    try:

        if choice == "1":

            print("Balance:", balance)

        elif choice == "2":

            try:

                amount = int(input("Amount: "))

                if amount > balance:

                    raise InsufficientBalance(
                        "Insufficient Balance"
                    )

                balance -= amount

                print("Withdrawal Successful")

            except ValueError:

                print("Enter Numbers Only")

        elif choice == "3":

            break

        else:

            print("Invalid Choice")

    except InsufficientBalance as e:

        print(e)

    finally:

        print("Transaction Completed")
````
