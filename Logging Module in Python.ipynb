# 📝 Logging Module in Python

---

# 1️⃣ What is Logging?

## Definition

**Logging** is the process of recording events that happen while a program is running.

These events may include:

* Information messages
* Warnings
* Errors
* Critical failures
* Debugging details

---

## Simple Meaning

Logging is like writing a diary of everything happening inside a program.

Instead of using many `print()` statements, professional applications use logging.

---

# 2️⃣ Real-World Analogy

## 👮 Security Guard Register

Imagine a security guard maintaining a register.

```text
Visitor entered      → Information
Unknown person       → Warning
Fight happened       → Error
Building on fire     → Critical
```

Everything is recorded for future reference.

👉 Logging works exactly like this.

**Logging = Program's Diary**

# 5️⃣ What is basicConfig()? (Very Important)

## Definition

`basicConfig()` is used to configure how logging should work.

It defines:

* Where logs are stored
* Which messages are shown
* How logs are formatted

---

## Real-World Analogy

📹 CCTV Camera Setup

Before recording, we decide:

```text
Where to save recordings?
Which events to record?
How recordings should appear?
```

Similarly, `basicConfig()` sets rules for logging.

---

## Syntax

```python
import logging

logging.basicConfig(
    filename="app.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
```

---

# 6️⃣ Parameters of basicConfig()

| Parameter | Purpose                     |
| --------- | --------------------------- |
| filename  | Store logs in a file        |
| level     | Minimum level to display    |
| format    | Format of log message       |
| filemode  | Write or append mode        |
| datefmt   | Custom date and time format |

---

## Example

```python
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(levelname)s - %(message)s"
)

logging.info("Program Started")
```

### Output

```text
INFO - Program Started
```

---

# 7️⃣ Understanding Logging Levels

Every log message has an importance level.

---

## Logging Levels Table

| Level    | Numeric Value | Purpose             |
| -------- | ------------- | ------------------- |
| CRITICAL | 50            | Very serious issue  |
| ERROR    | 40            | Program error       |
| WARNING  | 30            | Warning message     |
| INFO     | 20            | General information |
| DEBUG    | 10            | Detailed debugging  |
| NOTSET   | 0             | No level assigned   |

---

# 8️⃣ Real-World Analogy for Levels

## 🏥 Hospital Emergency

| Level    | Example                 |
| -------- | ----------------------- |
| CRITICAL | Patient critical        |
| ERROR    | Serious problem         |
| WARNING  | Minor issue             |
| INFO     | Patient stable          |
| DEBUG    | Detailed medical report |

---

# 9️⃣ Important Concept: Log Filtering

Higher levels automatically include more serious messages.

Example:

```python
import logging

logging.basicConfig(level=logging.WARNING)
```

Only these will be shown:

```text
WARNING
ERROR
CRITICAL
```

These will be ignored:

```text
DEBUG
INFO
```

---

## Visualization

```text

INFO (20)
WARNING (30)
ERROR (40)
CRITICAL (50)
```

If level is WARNING (30):

```text
Show:
WARNING
ERROR
CRITICAL

Hide:
DEBUG
INFO
```

# 1️⃣1️⃣ Logging Functions

Python provides these main logging functions.

| Function           | Purpose              |
| ------------------ | -------------------- |
| logging.debug()    | Detailed information |
| logging.info()     | General information  |
| logging.warning()  | Warning messages     |
| logging.error()    | Error messages       |
| logging.critical() | Serious failures     |

---

# 1️⃣2️⃣ Basic Example

```python
import logging

logging.basicConfig(level=logging.DEBUG)

logging.debug("Debug message")
logging.info("Info message")
logging.warning("Warning message")
logging.error("Error message")
logging.critical("Critical message")
```

### Output

```text
DEBUG:root:Debug message
INFO:root:Info message
WARNING:root:Warning message
ERROR:root:Error message
CRITICAL:root:Critical message
```

---

# 1️⃣3️⃣ What is root?

Notice this output:

```text
INFO:root:Message
```

`root` is the default logger created by Python.

It is called the **Root Logger**.

# 1️⃣4️⃣ Custom Message Format

We can control how logs look.

---

## Format Codes

| Format        | Meaning        |
| ------------- | -------------- |
| %(asctime)s   | Date and Time  |
| %(levelname)s | Log Level      |
| %(message)s   | Actual Message |
| %(name)s      | Logger Name    |

---

## Example

```python
import logging

logging.basicConfig(
    format="%(asctime)s - %(levelname)s - %(message)s",
    level=logging.INFO
)

logging.info("Program Started")
```

### Output

```text
2026-04-13 18:30:45 - INFO - Program Started
```

---

# 1️⃣5️⃣ Date & Time Formatting

Use `datefmt`.

```python
import logging

logging.basicConfig(
    format="%(asctime)s - %(message)s",
    datefmt="%d-%m-%Y %I:%M:%S %p",
    level=logging.INFO
)

logging.info("Program Started")
```

---

### Output

```text
13-04-2026 06:30:45 PM - Program Started
```

---

# 1️⃣6️⃣ Date Format Codes

| Code | Meaning        |
| ---- | -------------- |
| %d   | Day            |
| %m   | Month          |
| %Y   | Year           |
| %I   | Hour (12-hour) |
| %H   | Hour (24-hour) |
| %M   | Minutes        |
| %S   | Seconds        |
| %p   | AM/PM          |

# 1️⃣7️⃣ Logging to a File

Logs can be stored in a file.

```python
import logging

logging.basicConfig(
    filename="app.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logging.info("Program Started")
logging.error("Error Occurred")
```

---

## Stored In

```text
app.log
```

Example Content:

```text
2026-04-13 06:30:45 - INFO - Program Started
2026-04-13 06:31:00 - ERROR - Error Occurred
```

# 2️⃣1️⃣ Logging vs Print

| Feature      | print()        | logging       |
| ------------ | -------------- | ------------- |
| Purpose      | Display output | Record events |
| Levels       | ❌ No           | ✅ Yes         |
| Save to file | ❌ No           | ✅ Yes         |
| Professional | ❌ No           | ✅ Yes         |
| Debugging    | Limited        | Excellent     |

---

## Example

Using Print:

```python
print("Program Started")
```

Using Logging:

```python
logging.info("Program Started")
```

Logging is preferred in professional applications.

✅ Default logging level = WARNING

✅ DEBUG is used during development

✅ ERROR and CRITICAL are important in production

✅ Logging is more powerful than print()

✅ Logs can be stored in files

# 2️⃣3️⃣ Common Mistakes

## ❌ Forgetting basicConfig()

```python
logging.info("Hello")
```

May not show output because default level is WARNING.

---

## ❌ Using print() for debugging

```python
print("Value =", x)
```

Better:

```python
logging.debug("Value = %s", x)
```

---

## ❌ Using DEBUG in production

Too many unnecessary logs.
