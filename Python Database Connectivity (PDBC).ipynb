# Chapter 1: Introduction to PDBC

## What is PDBC?

### Definition

**Python Database Connectivity (PDBC)** is a technique that allows Python programs to connect with databases (such as **MySQL**) and perform database operations like storing, retrieving, updating, and deleting data.

---

### Simple Meaning

PDBC creates a communication bridge between **Python** and a **Database**.

```
Python  ⇄  MySQL Database
```

Python sends SQL queries to the database, and the database returns the results.

---

### One-Line Definition

> **PDBC is used to connect Python applications with databases and perform CRUD (Create, Read, Update, Delete) operations.**

---

# Chapter 2: Why Do We Need PDBC?

Without a database:

- Data is lost after the program closes.
- Large amounts of data cannot be managed efficiently.
- Searching and updating data becomes difficult.

With a database:

- Data is stored permanently.
- Fast searching and retrieval.
- Easy updating and deletion.
- Supports multiple users.

# Chapter 5: Basic Steps in PDBC

Every Python database program follows the same sequence.

```
1. Import mysql.connector

↓

2. Connect to Database

↓

3. Create Cursor

↓

4. Execute SQL Query

↓

5. Fetch Results (Optional)

↓

6. Commit Changes (For INSERT/UPDATE/DELETE)

↓

7. Close Cursor

↓

8. Close Connection
```

# Chapter 7: Creating a Cursor (`cursor()`)

## Definition

A **Cursor** is an object used to execute SQL queries and retrieve results from the database.

---

### One-Line Definition

> **Cursor acts as a bridge between Python and the database.**

---

## Example

```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="yogi",
    database="college"
)

cursor = conn.cursor()

print("Cursor Created")

conn.close()
```

---

# Chapter 8: Executing SQL Queries (`execute()`)

## Definition

Executes a single SQL statement.

---

## Create Table

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students(
    id INT PRIMARY KEY,
    name VARCHAR(50)
)
""")
```

---

## Insert Data

```python
cursor.execute(
    "INSERT INTO students VALUES (%s,%s)",
    (1,"Yogi")
)

conn.commit()
```

---

### Why Use `%s`?

Using parameterized queries prevents **SQL Injection** and automatically handles data formatting.

Avoid:

```python
cursor.execute("INSERT INTO students VALUES (1,'Yogi')")
```

Prefer:

```python
cursor.execute(
    "INSERT INTO students VALUES (%s,%s)",
    (1,"Yogi")
)
```

---

# Chapter 9: Executing Multiple Queries (`executemany()`)

## Definition

Executes the same SQL query multiple times with different values.

---

## Example

```python
students = [
    (2,"Ram"),
    (3,"Sam"),
    (4,"Amit")
]

cursor.executemany(
    "INSERT INTO students VALUES (%s,%s)",
    students
)

conn.commit()
```

---

## Advantages

- Faster
- Cleaner
- Better performance

# Chapter 11: Fetching Data

---

# 11.1 `fetchone()`

## Definition

Returns one row at a time.

---

```python
cursor.execute("SELECT * FROM students")

row = cursor.fetchone()

print(row)
```

Example Output

```
(1, 'Yogi')
```

---

Reading one row at a time

```python
while True:

    row = cursor.fetchone()

    if row is None:
        break

    print(row)
```

---

# 11.2 `fetchall()`

## Definition

Returns all rows.

---

```python
cursor.execute("SELECT * FROM students")

rows = cursor.fetchall()

print(rows)
```

Output

```
[(1,'Yogi'),
 (2,'Ram'),
 (3,'Sam')]
```

---

Loop Example

```python
cursor.execute("SELECT * FROM students")

for row in cursor.fetchall():
    print(row)
```

---

# 11.3 `fetchmany()`

## Definition

Returns a specified number of rows.

---

```python
cursor.execute("SELECT * FROM students")

rows = cursor.fetchmany(2)

print(rows)
```

Output

```
[(1,'Yogi'),
 (2,'Ram')]
```

---

# Comparison of Fetch Methods

| Method | Returns |
|----------|----------|
| `fetchone()` | One row |
| `fetchmany(n)` | First n rows |
| `fetchall()` | All rows |

# Chapter 12: Saving Changes (`commit()`)

## Definition

Permanently saves database changes.

---

Without `commit()`,

changes are **not stored permanently**.

---

Example

```python
cursor.execute(
    "INSERT INTO students VALUES (%s,%s)",
    (10,"Raj")
)

conn.commit()
```

---

Used with

- INSERT
- UPDATE
- DELETE

---

# Chapter 13: Undo Changes (`rollback()`)

## Definition

Cancels changes made during the current transaction before they are committed.

---

Example

```python
cursor.execute(
    "INSERT INTO students VALUES (%s,%s)",
    (20,"Temp")
)

conn.rollback()
```

No row is inserted because the transaction is rolled back.

---

## Real-Life Analogy

Imagine filling a form.

Before submitting,

you press **Cancel**.

The changes disappear.

That is what `rollback()` does.

---

# Chapter 16: Error Handling (Best Practice)

Always use `try...except`.

```python
import mysql.connector

try:

    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="yogi",
        database="college"
    )

    cursor = conn.cursor()

    cursor.execute("SELECT * FROM students")

    for row in cursor.fetchall():
        print(row)

except mysql.connector.Error as err:

    print("Database Error:", err)

finally:

    if 'cursor' in locals():
        cursor.close()

    if 'conn' in locals() and conn.is_connected():
        conn.close()
```

# Chapter 19: Best Practices

✅ Always use parameterized queries (`%s`).

✅ Always call `commit()` after INSERT, UPDATE, or DELETE.

✅ Use `rollback()` when a transaction fails.

✅ Close both cursor and connection.

✅ Handle exceptions using `try...except`.

✅ Never hardcode sensitive credentials in production code. Consider using environment variables or configuration files.

# Interview Questions

### 1. What is PDBC?

PDBC (Python Database Connectivity) is a technique used to connect Python applications with databases and perform CRUD operations.

---

### 2. Why is a cursor required?

A cursor is used to execute SQL queries and retrieve results from the database.

---

### 3. What is the difference between `execute()` and `executemany()`?

- `execute()` executes a single SQL statement.
- `executemany()` executes the same SQL statement multiple times with different values.

---

### 4. Why is `commit()` necessary?

`commit()` permanently saves INSERT, UPDATE, and DELETE operations to the database.

---

### 5. What is `rollback()`?

`rollback()` cancels the current transaction before it is committed, restoring the database to its previous state.

---

### 6. What is the difference between `fetchone()`, `fetchmany()`, and `fetchall()`?

- `fetchone()` returns one row.
- `fetchmany(n)` returns the next `n` rows.
- `fetchall()` returns all remaining rows.

---

### 7. Why should parameterized queries (`%s`) be used?

They help prevent SQL Injection attacks and safely pass user data to SQL queries.

---

### 8. Why should connections and cursors always be closed?

Closing them releases database resources, prevents memory leaks, and avoids exhausting available database connections.

# Real-World Teaching Line

Whenever a Python program communicates with a database:

1. Connect to the database.
2. Create a cursor.
3. Execute SQL queries.
4. Fetch results if needed.
5. Commit changes for INSERT, UPDATE, or DELETE.
6. Roll back if an error occurs during a transaction.
7. Close the cursor and the connection.

Following these steps ensures that database applications are secure, efficient, and reliable.
