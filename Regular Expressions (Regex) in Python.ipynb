# 6️⃣ Regex Match Objects

Many regex functions return a **Match Object**.

A Match Object contains:

* Matched value
* Start position
* End position


# 7️⃣ Important Match Methods

## group()

Returns matched text.

```python
import re

m = re.search("abc", "xyzabc123")

print(m.group())
```

Output:

```text
abc
```

---

## start()

Returns starting index.

```python
print(m.start())
```

Output:

```text
3
```

---

## end()

Returns ending index.

```python
print(m.end())
```

Output:

```text
6
```

---

## span()

Returns both start and end.

```python
print(m.span())
```

Output:

```text
(3, 6)
```

---

# 8️⃣ finditer()

## Definition

Returns matches one by one as Match Objects.

Useful when we need:

* Match value
* Start position
* End position

---

## Example

```python
import re

data = "abc123abc"

for m in re.finditer("abc", data):
    print(m.group(), m.start(), m.end())
```

Output:

```text
abc 0 3
abc 6 9
```

# 9️⃣ Character Classes

Character classes match groups of characters.

---

## [ABC]

Matches:

```text
A or B or C
```

```python
import re

print(re.findall("[ABC]", "ABCDXYZ"))
```

Output:

```python
['A', 'B', 'C']
```

---

## [A-Z]

Matches uppercase letters.

```python
print(re.findall("[A-Z]", "AbC123"))
```

Output:

```python
['A', 'C']
```

---

## [a-z]

Matches lowercase letters.

```python
print(re.findall("[a-z]", "AbC123"))
```

Output:

```python
['b']
```

---

## [0-9]

Matches digits.

```python
print(re.findall("[0-9]", "abc123"))
```

Output:

```python
['1', '2', '3']
```

---

## [A-Za-z]

Matches all alphabets.

```python
print(re.findall("[A-Za-z]", "abc123XYZ"))
```

Output:

```python
['a','b','c','X','Y','Z']
```

---

## [A-Za-z0-9]

Matches letters and digits.

```python
print(re.findall("[A-Za-z0-9]", "@abc123#"))
```

Output:

```python
['a','b','c','1','2','3']
```

---

## [^A-Za-z0-9]

Matches everything except letters and digits.

```python
print(re.findall("[^A-Za-z0-9]", "@abc123#"))
```

Output:

```python
['@', '#']
```

# 1️⃣1️⃣ Quantifiers

Quantifiers tell how many times a character should occur.

---

## a

Exactly one

```python
print(re.findall("a", "aaa"))
```

Output:

```python
['a','a','a']
```

---

## a+

One or more

```python
print(re.findall("a+", "aaab"))
```

Output:

```python
['aaa']
```

---

## a*

Zero or more

```python
print(re.findall("a*", "aaab"))
```

---

## a?

Zero or one

```python
print(re.findall("a?", "abc"))
```

---

## a{3}

Exactly 3

```python
print(re.findall("a{3}", "aaabaaa"))
```

Output:

```python
['aaa', 'aaa']
```

---

## a{2,6}

Between 2 and 6

```python
print(re.findall("a{2,6}", "aaaaaaab"))
```

Output:

```python
['aaaaaa']
```

# 1️⃣4️⃣ Important Regex Functions

## match()

Checks only at beginning.

```python
print(re.match("abc", "abcdef"))
```

---

## fullmatch()

Entire string must match.

```python
print(re.fullmatch("abc", "abc"))
```

---

## search()

Search anywhere.

```python
print(re.search("abc", "xyzabc"))
```

---

## findall()

Returns all matches.

```python
print(re.findall("a", "banana"))
```

Output:

```python
['a','a','a']
```

---

## sub()

Replace matches.

```python
print(re.sub("a", "X", "banana"))
```

Output:

```text
bXnXnX
```

---

## subn()

Replace + count.

```python
print(re.subn("a", "X", "banana"))
```

Output:

```python
('bXnXnX', 3)
```

# 🔍 Why Do We Use `r""` (Raw Strings) in Python Regex?

---

# 1️⃣ Introduction

When learning Regular Expressions (Regex), you will often see patterns written like:

```python
r"\d"
r"\w"
r"\s"
r"\W"
```

Many students ask:

> Why do we write `r` before the string?

To understand this, we first need to understand **escape characters**.

---

# 2️⃣ What is an Escape Character?

In Python, the backslash (`\`) is a special character.

It is used to create **escape sequences**.

---

## Common Escape Sequences

| Escape Sequence | Meaning      |
| --------------- | ------------ |
| `\n`            | New Line     |
| `\t`            | Tab Space    |
| `\\`            | Backslash    |
| `\'`            | Single Quote |
| `\"`            | Double Quote |

---

## Example

```python
print("Hello\nWorld")
```

### Output

```text
Hello
World
```

### Explanation

Python sees:

```python
\n
```

and converts it into a new line.

---

# 3️⃣ Problem in Regex

Regex also uses backslashes.

Example:

```python
\d
\w
\s
```

But Python may think these are escape sequences instead of regex patterns.

This creates confusion.

---

# 4️⃣ Solution: Raw Strings

Python provides **Raw Strings**.

A raw string is created by placing:

```python
r
```

before the string.

---

## Syntax

```python
r"pattern"
```

Example:

```python
r"\d"
```

---

## Definition

A raw string tells Python:

> Treat every character exactly as written.

Do not process backslashes as escape sequences.

---

# 5️⃣ Example Without Raw String

```python
print("Hello\nWorld")
```

### Output

```text
Hello
World
```

Python interprets:

```python
\n
```

as a newline.

---

# 6️⃣ Example With Raw String

```python
print(r"Hello\nWorld")
```

### Output

```text
Hello\nWorld
```

### Explanation

Python prints:

```python
\n
```

exactly as written.

Because:

```python
r""
```

prevents escape sequence processing.

---

# 7️⃣ Why Raw Strings are Important in Regex

Regex patterns contain many backslashes:

```python
\d
\w
\s
\b
\D
\W
\S
```

Using raw strings makes patterns easier to read.

---

## Recommended

```python
r"\d"
```

---

## Equivalent But Harder to Read

```python
"\\d"
```

Both work.

But:

```python
r"\d"
```

is cleaner and easier.

---

# 8️⃣ Comparison

## Without Raw String

```python
pattern = "\\d"
```

---

## With Raw String

```python
pattern = r"\d"
```

Both mean the same thing.

Most programmers prefer:

```python
r"\d"
```

because it is easier to understand.

# 🔟 Understanding `\w`

## Definition

`\w` matches:

```text
A-Z
a-z
0-9
_
```

---

## Example

```python
import re

print(re.findall(r"\w", "@abc123"))
```

### Output

```python
['a', 'b', 'c', '1', '2', '3']
```

---

## Explanation

String:

```text
@abc123
```

Checking characters:

| Character | Word Character? |
| --------- | --------------- |
| @         | ❌ No            |
| a         | ✅ Yes           |
| b         | ✅ Yes           |
| c         | ✅ Yes           |
| 1         | ✅ Yes           |
| 2         | ✅ Yes           |
| 3         | ✅ Yes           |

Only word characters are returned.

---

# 1️⃣1️⃣ Understanding `\W`

## Definition

`\W` means:

> Match any character that is NOT a word character.

---


```python

```

---

## Explanation

String:

```text

```

| Character | Word Character? |
| --------- | --------------- |
| @         | ❌ No            |
| a         | ✅ Yes           |
| b         | ✅ Yes           |
| c         | ✅ Yes           |
| 1         | ✅ Yes           |
| 2         | ✅ Yes           |
| 3         | ✅ Yes           |

Since `@` is not a word character:

```python
['@']
```

is returned.

# 🎯 Interview Question

### Q. Why do we use `r""` in Regex?

**Answer:**

Raw strings (`r""`) tell Python not to treat backslashes as escape characters. This allows Regex patterns like `\d`, `\w`, and `\s` to be written clearly and correctly without using extra backslashes.

---

# 🎓 1-Minute Classroom Explanation

"Python normally treats backslashes as escape characters. For example, `\n` means a new line and `\t` means a tab. But Regex patterns also use backslashes, such as `\d` for digits and `\w` for word characters. To prevent Python from interpreting these backslashes as escape sequences, we use raw strings by placing `r` before the string. A raw string tells Python to use the text exactly as written. That's why most Regex patterns are written as `r'\d'`, `r'\w'`, and `r'\s'`."

---

# 🚀 Final One-Line Summary

**`r""` creates a raw string, telling Python to treat backslashes literally, which makes Regex patterns easier to write, read, and understand.**
