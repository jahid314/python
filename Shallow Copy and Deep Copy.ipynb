# Chapter 1: Introduction

When working with Python, we often need to create a copy of an object such as a list, dictionary, or other collections.

However, **not all copies are the same**.

Python provides different ways to copy objects:

1. Assignment (`=`)
2. Shallow Copy (`copy.copy()`)
3. Deep Copy (`copy.deepcopy()`)

Understanding the difference between these is very important because it affects how changes to one object impact another.

---

# Chapter 2: Why is Copy Needed?

Imagine you have a list:

```python
a = [10, 20, 30]
```

Now you assign it to another variable:

```python
b = a
```

Many beginners think Python creates another list.

**But that's not true.**

Python only creates another reference (another name) for the same object.

---

## Example

```python
a = [1, 2, 3]

b = a

print(a)
print(b)
```

### Output

```
[1, 2, 3]
[1, 2, 3]
```

---

## Check Memory Address

```python
print(id(a))
print(id(b))
```

### Output

```
140452356789120
140452356789120
```

(Both addresses are the same.)

---

## Observation

- `a` and `b` point to the **same object**.
- No new list is created.

---

## Modify One Variable

```python
b[0] = 100

print(a)
print(b)
```

### Output

```
[100, 2, 3]
[100, 2, 3]
```

---

## Why?

Because both variables refer to the **same memory location**.

Changing one changes the other.

---

## Memory Diagram

```
        +----------------+
a ----> | [1, 2, 3]      |
        +----------------+
             ^
             |
b -----------+
```

---

## Real-World Analogy

Imagine one notebook.

Two friends call it their notebook.

Friend A writes something.

Friend B immediately sees the change because there is only **one notebook**.

---

# Chapter 3: Understanding Nested Lists

Before learning shallow and deep copy, understand nested lists.

Example

```python
a = [[1, 2], [3, 4]]
```

This is **not one list only**.

It is

- One outer list
- Two inner lists

Memory representation

```
Outer List
   |
   |------> [1, 2]
   |
   |------> [3, 4]
```

This concept is important because shallow copy only copies the outer list.

---

# Chapter 4: Shallow Copy

## Definition

A **Shallow Copy** creates a **new outer object**, but the **inner objects are shared** between the original and the copied object.

---

## One-Line Definition

> **Shallow Copy creates a new outer object while sharing references to nested objects.**

---

## Import Copy Module

```python
import copy
```

---

## Syntax

```python
copy.copy(object)
```

---

## Example

```python
import copy

a = [[1, 2], [3, 4]]

b = copy.copy(a)
```

---

## Check Addresses

```python
print("Outer Address of a:", id(a))
print("Outer Address of b:", id(b))

print("Inner Address of a:", id(a[0]))
print("Inner Address of b:", id(b[0]))
```

### Example Output

```
Outer Address of a: 140251900
Outer Address of b: 140251980

Inner Address of a: 140251300
Inner Address of b: 140251300
```

---

## Observation

### Outer List

Different memory addresses.

### Inner Lists

Same memory addresses.

---

## Memory Diagram

```
Original (a)

+---------------------------+
|     Outer List            |
|  +---------+              |
|  | [1,2]   |--------------+
|  +---------+              |
|  | [3,4]   |--------------+
|  +---------+              |
+---------------------------+

Copied (b)

+---------------------------+
|     Outer List            |
|  +---------+--------------+
|  | [1,2]   |
|  +---------+--------------+
|  | [3,4]   |
|  +---------+--------------+
+---------------------------+
```

Only the outer list is copied.

Inner lists are shared.

---

## Modify Inner List

```python
b[0][0] = 100

print(a)
print(b)
```

### Output

```
[[100, 2], [3, 4]]

[[100, 2], [3, 4]]
```

---

## Why?

Because both variables point to the same inner list.

---

## Modify Outer List Only

```python
b.append([5, 6])

print(a)
print(b)
```

### Output

```
[[1, 2], [3, 4]]

[[1, 2], [3, 4], [5, 6]]
```

Appending affects only the copied outer list.

---

## Important Points

✔ Outer object is copied.

✔ Nested objects are shared.

✔ Changes inside nested objects affect both variables.

✔ Changes to the outer list affect only the copied list.

---

## Real-World Analogy

Imagine making a photocopy of a file.

The file cover is new.

But the attached documents are still the same originals.

Changing an attachment changes it for both files.

---

# Chapter 5: Deep Copy

## Definition

A **Deep Copy** creates a completely independent copy of the object, including all nested objects.

---

## One-Line Definition

> **Deep Copy creates entirely new copies of both outer and inner objects.**

---

## Syntax

```python
copy.deepcopy(object)
```

---

## Example

```python
import copy

a = [[1, 2], [3, 4]]

b = copy.deepcopy(a)
```

---

## Check Addresses

```python
print(id(a))
print(id(b))

print(id(a[0]))
print(id(b[0]))
```

### Output

```
Outer Address → Different

Inner Address → Different
```

---

## Observation

Everything is copied independently.

---

## Memory Diagram

```
Original

Outer List

    |
    |----> [1,2]
    |
    |----> [3,4]

Copied

Outer List

    |
    |----> [1,2]
    |
    |----> [3,4]
```

No object is shared.

---

## Modify Copied Object

```python
b[0][0] = 100

print(a)
print(b)
```

### Output

```
[[1, 2], [3, 4]]

[[100, 2], [3, 4]]
```

---

## Why?

Because every object has been copied independently.

---

## Important Points

✔ Outer object is new.

✔ Inner objects are new.

✔ Nothing is shared.

✔ Safe for nested structures.

---

## Real-World Analogy

Imagine making a completely new file.

You also create brand-new copies of every attached document.

Changing one file never affects the original.

# Chapter 6: Assignment vs Shallow Copy vs Deep Copy

## Address Comparison

| Method | Outer Address | Inner Address |
|---------|---------------|---------------|
| Assignment (`=`) | Same | Same |
| Shallow Copy (`copy.copy()`) | Different | Same |
| Deep Copy (`copy.deepcopy()`) | Different | Different |

---

## Modification Behavior

| Operation | Assignment | Shallow Copy | Deep Copy |
|------------|------------|--------------|-----------|
| Change outer list | Affects both | Affects only copied list | Affects only copied list |
| Change inner list | Affects both | Affects both | Affects only copied list |

---

# Chapter 7: Another Way to Create a Shallow Copy

## Using List Slicing

```python
a = [[1, 2], [3, 4]]

b = a[:]
```

---

## Check Addresses

```python
print(id(a))
print(id(b))

print(id(a[0]))
print(id(b[0]))
```

### Output

```
Outer → Different

Inner → Same
```

---

## Modify Inner List

```python
b[0][0] = 100

print(a)
print(b)
```

Output

```
[[100, 2], [3, 4]]

[[100, 2], [3, 4]]
```

This is still a **Shallow Copy**.

---

# Chapter 8: Dictionary Example

Shallow Copy

```python
import copy

student = {
    "marks": [90, 80]
}

student2 = copy.copy(student)

student2["marks"][0] = 100

print(student)
print(student2)
```

Output

```
{'marks': [100, 80]}
{'marks': [100, 80]}
```

---

Deep Copy

```python
import copy

student = {
    "marks": [90, 80]
}

student2 = copy.deepcopy(student)

student2["marks"][0] = 100

print(student)
print(student2)
```

Output

```
{'marks': [90, 80]}
{'marks': [100, 80]}
```

# Chapter 9: When Should You Use Each?

## Use Assignment (`=`)

When you intentionally want two variables to refer to the same object.

---

## Use Shallow Copy

When:

- The object contains only simple (non-nested) values.
- You only need a new outer object.

---

## Use Deep Copy

When:

- Working with nested lists.
- Working with nested dictionaries.
- Copying complex objects.
- You need complete independence between the original and the copy.

---

# Chapter 10: Common Mistakes

### Mistake 1

Thinking

```python
b = a
```

creates a new object.

❌ Wrong.

It only creates another reference.

---

### Mistake 2

Using shallow copy for nested data and expecting complete independence.

❌ Wrong.

Nested objects are still shared.

---

### Mistake 3

Forgetting to import the `copy` module.

```python
import copy
```

is required for `copy.copy()` and `copy.deepcopy()`.

# Interview Questions

### 1. What is the difference between Assignment and Shallow Copy?

Assignment creates another reference to the same object, while Shallow Copy creates a new outer object but shares nested objects.

---

### 2. What is Deep Copy?

Deep Copy creates completely independent copies of both the outer object and all nested objects.

---

### 3. Why is the `copy` module required?

The `copy` module provides the `copy()` and `deepcopy()` functions used to create shallow and deep copies.

---

### 4. Does list slicing (`[:]`) create a Deep Copy?

No.

It creates a **Shallow Copy**.

---

### 5. When should you use `deepcopy()`?

When working with nested lists, nested dictionaries, or any complex object where changes to the copied object should not affect the original.

---

### 6. Which copy method is safest for nested data structures?

`copy.deepcopy()` because it creates completely independent copies of every nested object.

---

# Real-World Teaching Line

Before copying an object, ask yourself:

- Do I want both variables to share the same data? → Use **Assignment (`=`)**.
- Do I only need a new outer container? → Use **Shallow Copy (`copy.copy()` or `[:]`)**.
- Do I need a completely independent copy, including all nested objects? → Use **Deep Copy (`copy.deepcopy()`)**.

Choosing the correct method helps prevent unexpected bugs caused by shared references.

---

# Final One-Line Summary

> **Assignment creates another reference to the same object, Shallow Copy creates a new outer object but shares nested objects, while Deep Copy creates completely independent copies of both outer and inner objects, making it the safest option for nested data structures.**
