# 1. What is CSV?

## Definition

CSV stands for:

```text
Comma Separated Values
```

It is a file format used to store tabular data in rows and columns.

---

## Real-World Analogy

Imagine an Excel sheet:

| Name | Marks |
| ---- | ----- |
| Ravi | 90    |
| Yogi | 95    |

CSV stores it like:

```csv
Name,Marks
Ravi,90
Yogi,95
```

# 5. Writing CSV Files

## Syntax

```python
import csv

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Marks"])
    writer.writerow(["Yogi", 95])
```

---

## Why newline="" ?

Always use:

```python
newline=""
```

Without it, Windows may add blank rows.

Correct:

```python
open("file.csv", "w", newline="")
```

---

## Example

```python
import csv

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Marks"])
    writer.writerow(["Ravi", 85])
    writer.writerow(["Yogi", 95])

print("Data Written Successfully")
```

# 6. Reading CSV Files

# 7. writerow() vs writerows()

---

## writerow()

Writes one row.

```python
writer.writerow(["Ravi", 90])
```

---

## writerows()

Writes multiple rows.

```python
rows = [
    ["Ravi", 90],
    ["Yogi", 95],
    ["Sai", 88]
]

writer.writerows(rows)
```

---

## Complete Example

```python
import csv

rows = [
    ["Ravi", 90],
    ["Yogi", 95],
    ["Sai", 88]
]

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Marks"])
    writer.writerows(rows)
```

# 9. DictWriter

Instead of lists, use dictionaries.

---

## Example

```python
import csv

fields = ["name", "marks"]

with open("students.csv", "w", newline="") as file:

    writer = csv.DictWriter(
        file,
        fieldnames=fields
    )

    writer.writeheader()

    writer.writerow(
        {
            "name": "Yogi",
            "marks": 95
        }
    )
```

---

## Output

```csv
name,marks
Yogi,95
```

# 14. Updating CSV Records

CSV files cannot directly update records.

Steps:

1. Read all data
2. Modify data
3. Rewrite file

---

## Example

```python
import csv

data = []

with open("students.csv", "r") as file:

    reader = csv.reader(file)

    for row in reader:

        if row[0] == "Ravi":
            row[1] = "100"

        data.append(row)

with open("students.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerows(data)
```

---

# 15. Deleting CSV Records

---

## Example

```python
import csv

data = []

name = input(
    "Enter name to delete: "
)

with open("students.csv", "r") as file:

    reader = csv.reader(file)

    for row in reader:

        if row[0] != name:
            data.append(row)

with open("students.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerows(data)
```
