# Chapter 1: Introduction to Multithreading

## What is Multithreading?

### Definition

**Multithreading** is a technique in which multiple threads execute within the same program, allowing multiple tasks to run concurrently.

---

### Simple Meaning

> **One program performs multiple tasks at the same time.**

Instead of waiting for one task to finish before starting another, Python can run multiple tasks concurrently using threads.

---

### One-Line Definition

> **Multithreading allows multiple threads to execute concurrently within a single process.**

---

## Real-World Analogy

Imagine a chef in a restaurant.

Instead of doing one task at a time:

- Boiling rice 🍚
- Frying vegetables 🍳
- Cutting onions 🧅

The chef performs these tasks together.

This saves time and improves efficiency.

Similarly,

A Python program can perform multiple tasks using threads.

---

# Chapter 2: Why Do We Need Multithreading?

Some tasks spend most of their time waiting.

Examples:

- Downloading files
- Uploading images
- Reading files
- Waiting for network responses
- Waiting for database results

Instead of waiting,

another thread can continue working.

This improves the responsiveness of the application.

# Chapter 4: What is a Thread?

## Definition

A **Thread** is the smallest unit of execution inside a process.

A process may contain one or many threads.

---

## Simple Meaning

Think of a process as a company.

Inside the company,

multiple employees work simultaneously.

- Company → Process
- Employees → Threads

---

# Chapter 5: Process vs Thread

| Process | Thread |
|----------|---------|
| Heavyweight | Lightweight |
| Separate memory | Shared memory |
| Slower creation | Faster creation |
| Communication is slower | Communication is faster |
| More resources | Less resources |

---

## Real-World Analogy

### Process

Different shops in a shopping mall.

Each shop has:

- Separate workers
- Separate inventory
- Separate electricity

---

### Thread

Employees inside one shop.

They share:

- Same inventory
- Same electricity
- Same workplace

---

# Chapter 6: Types of Multitasking

## 1. Process-Based Multitasking

### Definition

Multiple processes run simultaneously.

Each process has its own memory.

### Examples

- Chrome
- VS Code
- Spotify

All running together.

---

## Advantages

- Better isolation
- One crash doesn't affect others

---

## Disadvantages

- Higher memory usage
- Slower communication

---

## 2. Thread-Based Multitasking

### Definition

Multiple threads run inside one process.

Threads share the same memory.

### Examples

Inside YouTube:

- Download video
- Play video
- Load comments

All happen simultaneously.

---

## Advantages

- Fast communication
- Less memory
- Easy resource sharing

# Chapter 9: Function-Based Threading

## Syntax

```python
from threading import *

def task():
    print("Child Thread")

t = Thread(target=task)

t.start()
```

---

## Example 1

```python
from threading import *

def task():
    for i in range(3):
        print("Child Thread")

t = Thread(target=task)

t.start()

for i in range(3):
    print("Main Thread")
```

---

### Possible Output

```
Child Thread

Main Thread

Child Thread

Main Thread

Child Thread

Main Thread
```

---

## Important Note

The output order **may change** every time.

This is called **concurrent execution**.

---

# Chapter 10: Understanding `start()`

## Definition

Starts a new thread.

---

### Syntax

```python
thread.start()
```

---

## Example

```python
from threading import *

def task():
    print("Thread Started")

t = Thread(target=task)

t.start()
```

---

## Important Difference

### Wrong

```python
t.run()
```

Runs like a normal function.

No new thread.

---

### Correct

```python
t.start()
```

Creates a new thread.

# Chapter 12: Inheritance-Based Threading

## Definition

Create your own thread by inheriting the `Thread` class.

---

## Example

```python
from threading import *

class MyThread(Thread):

    def run(self):

        for i in range(3):
            print("Child Thread")

t = MyThread()

t.start()
```

---

## Why Override `run()`?

When `start()` is called,

Python automatically executes the `run()` method.

# 15.2 `join()`

## Definition

Makes one thread wait until another thread finishes.

---

### Example

```python
from threading import *

import time

def task():

    for i in range(3):

        print("Child Thread")

        time.sleep(1)

t = Thread(target=task)

t.start()

t.join()

print("Main Thread")
```

---

### Output

```
Child Thread

Child Thread

Child Thread

Main Thread
```

---

## Why Use `join()`?

To ensure one task finishes before another starts.

Useful in:

- File processing
- Report generation
- Database updates

---

# Chapter 16: Daemon Threads

## What is a Daemon Thread?

### Definition

A **Daemon Thread** is a background thread that automatically terminates when the main program finishes.

---

### Simple Meaning

Daemon threads work only while the main program is running.

When the main program ends,

daemon threads also stop immediately.

---

### One-Line Definition

> **A daemon thread runs in the background and automatically stops when the main thread exits.**

---

## Real-World Analogy

Office:

- Employees → Main Thread
- Security Guard → Daemon Thread

When the office closes,

the security guard also leaves.

---

# Chapter 17: Default Thread Nature

| Thread | Default Type |
|---------|--------------|
| Main Thread | Non-Daemon |
| Child Thread | Non-Daemon |

By default,

threads finish their work completely.

---

# Chapter 18: Creating a Daemon Thread

```python
from threading import *

import time

def task():

    while True:

        print("Daemon Running")

        time.sleep(1)

t = Thread(target=task)

t.daemon = True

t.start()

print("Main Thread Ends")
```

---

### Output

```
Daemon Running

Main Thread Ends
```

Daemon thread automatically stops.

---

## Important Rule

Always set daemon status **before** calling `start()`.

Correct

```python
t.daemon = True

t.start()
```

Wrong

```python
t.start()

t.daemon = True
```

Raises

```
RuntimeError
```

---

# Chapter 19: Non-Daemon Thread

```python
from threading import *

import time

def task():

    for i in range(5):

        print("Child Thread")

        time.sleep(1)

t = Thread(target=task)

t.start()

print("Main Ends")
```

Output

```
Main Ends

Child Thread

Child Thread

Child Thread

...
```

The child thread continues even after the main thread reaches its end.

# Chapter 20: Daemon vs Non-Daemon

| Feature | Daemon Thread | Non-Daemon Thread |
|----------|---------------|-------------------|
| Runs in background | ✅ Yes | ❌ No |
| Stops when main exits | ✅ Yes | ❌ No |
| Completes work | ❌ Not guaranteed | ✅ Yes |
| Used for critical tasks | ❌ No | ✅ Yes |

```

Output

```

## Using `isDaemon()` (Older API)

```python

```

> **Note:** `isDaemon()` is deprecated in modern Python. Prefer using the `.daemon` attribute.

# Chapter 22: Parent–Child Rule

A child thread inherits the daemon status of its parent.

```python
from threading import *

def child():
    print(current_thread().daemon)

def parent():

    t2 = Thread(target=child)

    t2.start()

t1 = Thread(target=parent)

t1.daemon = True

t1.start()
```

Output

```
True
```

---

## Override Daemon Status

```python
t2.daemon = False
```

You can change it,

but **only before `start()`**.

# Chapter 26: Limitations of Multithreading in Python

Python has a mechanism called the **Global Interpreter Lock (GIL)**.

### What is the GIL?

The **Global Interpreter Lock (GIL)** allows only **one thread to execute Python bytecode at a time** within a single Python process.

### Effect of the GIL

- Excellent for **I/O-bound tasks** (networking, file handling, downloading).
- Does **not** provide true parallel execution for **CPU-bound tasks** (heavy mathematical calculations) in standard CPython.

For CPU-intensive work, Python's `multiprocessing` module is often a better choice.

# Interview Questions

### 1. What is multithreading?

Multithreading is a technique that allows multiple threads to execute concurrently within a single process.

---

### 2. What is the difference between a process and a thread?

A process has its own memory space, while threads within the same process share memory and resources.

---

### 3. What is the purpose of `start()`?

It creates a new thread and begins executing its `run()` method.

---

### 4. What is the purpose of `join()`?

It makes one thread wait until another thread completes its execution.

---

### 5. What is a daemon thread?

A daemon thread is a background thread that automatically stops when the main program exits.

---

### 6. Can you change the daemon status after calling `start()`?

No. The daemon status must be set **before** calling `start()`. Otherwise, Python raises a `RuntimeError`.

---

### 7. What does `active_count()` return?

It returns the total number of currently active threads.

---

### 8. What does `is_alive()` do?

It returns `True` if a thread is currently running; otherwise, it returns `False`.

---

### 9. What is the Global Interpreter Lock (GIL)?

The GIL is a mechanism in CPython that allows only one thread to execute Python bytecode at a time, making multithreading ideal for I/O-bound tasks but not for CPU-bound tasks.

---

# Real-World Teaching Line

Before using multithreading, ask yourself:

- Is the task waiting for input/output (network, files, database)? → **Use multithreading.**
- Is the task CPU-intensive (heavy calculations)? → Consider **multiprocessing** instead.
- Is the task critical (saving files, payments, database updates)? → **Use non-daemon threads** to ensure the work completes.

Choosing the right threading strategy improves both performance and reliability.

---

# Final One-Line Summary

> **Multithreading allows a single Python process to perform multiple tasks concurrently using threads. It is especially effective for I/O-bound operations, while daemon threads are background threads that automatically stop when the main program exits.**
