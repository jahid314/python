# 1️⃣ What is File Handling?

## Definition

**File Handling** is the process of:

- Creating files
- Reading data from files
- Writing data to files
- Updating file contents
- Deleting files

using a programming language.

In Python, File Handling allows programs to store data permanently on the computer.

---

## Example

Suppose a user enters:

```python
name = input("Enter your name: ")
print(name)
```

Output:

```text
Jahid
```

After the program closes:

❌ Data is lost

Because data was stored only in RAM.

---

Using File Handling:

```python
with open("data.txt", "w") as file:
    file.write("Jahid")
```

Now:

✅ Data is stored permanently

Even after the program closes.

---

# 3️⃣ How File Handling Works Internally

## System-Level Architecture

```text
Python Program
       ↓
Operating System
       ↓
File System
       ↓
Hard Disk / SSD
```

---

## What Happens When You Open a File?

Example:

```python
file = open("data.txt", "r")
```

Python performs:

### Step 1

Requests Operating System:

```text
Open data.txt
```

### Step 2

Operating System checks:

- File exists?
- Permission available?

### Step 3

OS creates a File Descriptor.

### Step 4

Python receives a File Object.

### Step 5

Program can now read/write data.

# 7️⃣ File Modes

---

## Read Mode (`r`)

Used to read existing files.

```python
file = open("data.txt", "r")
```

If file does not exist:

```python
FileNotFoundError
```

---

## Write Mode (`w`)

Used to write data.

```python
file = open("data.txt", "w")
```

### Important

- Creates file if not exists
- Deletes old content

Example:

Before:

```text
Hello
```

After:

```python
file.write("Python")
```

Result:

```text
Python
```

---

## Append Mode (`a`)

Adds data at the end.

```python
file = open("data.txt", "a")
```

Example:

Before:

```text
Hello
```

Append:

```python
file.write(" Python")
```

Result:

```text
Hello Python
```

---

## Create Mode (`x`)

Creates a new file.

```python
file = open("newfile.txt", "x")
```

If file already exists:

```python
FileExistsError
```

---

## Read and Write Mode (`r+`)

```python
file = open("data.txt", "r+")
```

Can:

- Read
- Write

---

## Write and Read Mode (`w+`)

```python
file = open("data.txt", "w+")
```

Can:

- Read
- Write
- Overwrites file

---

## Append and Read Mode (`a+`)

```python
file = open("data.txt", "a+")
```

Can:

- Append
- Read

# 8️⃣ Reading Files

---

## read()

Reads complete file.

```python
with open("data.txt", "r") as file:
    print(file.read())
```

Output:

```text
Hello Python
```

---

## read(size)

Reads specified number of characters.

```python
with open("data.txt", "r") as file:
    print(file.read(5))
```

Output:

```text
Hello
```

---

## readline()

Reads one line.

```python
with open("data.txt", "r") as file:
    print(file.readline())
```

---

## readlines()

Returns all lines as list.

```python
with open("data.txt", "r") as file:
    print(file.readlines())
```

Output:

```python
['Line1\n', 'Line2\n']
```

# 1️⃣1️⃣ Closing Files

## Why Close Files?

When a file is opened:

- OS allocates memory
- OS creates file descriptor

Resources remain occupied.

---

## Traditional Way

```python
file = open("data.txt", "r")

print(file.read())

file.close()
```

---

# Problem

Suppose:

```python
file = open("data.txt", "r")

10 / 0
```

Program crashes.

File remains open.

This causes:

- Resource leakage
- Memory wastage

---

# 1️⃣2️⃣ The `with` Statement (Professional Method)

## Why Use It?

Automatically closes files.

Recommended by Python developers.

---

## Syntax

```python
with open("data.txt", "r") as file:
    print(file.read())
```

---

## Internal Working

```text
Open File
    ↓
Execute Block
    ↓
Automatically Close File
```

Even if exception occurs:

```python
with open("data.txt", "r") as file:
    print(10/0)
```

File still closes properly.

# 1️⃣4️⃣ seek() and tell()

---

## tell()

Returns current cursor position.

```python
with open("data.txt", "r") as file:
    file.read(5)
    print(file.tell())
```

Output:

```python
5
```

---

## seek()

Moves cursor to desired location.

```python
with open("data.txt", "r") as file:
    file.seek(0)
```

Moves cursor to beginning.

---

## Example

```python
with open("data.txt", "r") as file:
    file.read(5)

    file.seek(0)

    print(file.read())
```

Output:

```text
Hello Python
```

# 1️⃣5️⃣ Exception Handling with Files

## Why Needed?

File may:

- Not exist
- Be locked
- Lack permissions

---

## Example

```python
try:
    with open("abc.txt", "r") as file:
        print(file.read())

except FileNotFoundError:
    print("File not found")
```

Output:

```text
File not found
```
