# 📦 Zipping & Unzipping in Python

---

# 1️⃣ What is Zipping?

## Definition
**Zipping** means compressing one or more files into a single `.zip` file.

A ZIP file can contain:
- One file
- Multiple files
- Entire folders

---

## Real-World Analogy

🎒 **School Bag Example**

| Real World | Computer World |
|------------|----------------|
| Books | Files |
| School Bag | ZIP File |
| Packing books into bag | Zipping files |

👉 Just like multiple books can be packed into one bag, multiple files can be packed into one ZIP file.

---

# 2️⃣ Why Do We Use Zipping?

### Advantages

✅ Saves storage space

✅ Easy to share files

✅ Combines multiple files into one file

✅ Faster file transfer

✅ Useful for backups

---

# 3️⃣ ZIP File Extension

A zipped file usually has the extension:

```text
.zip
```

Example:

```text
documents.zip
photos.zip
backup.zip
```

---

# 4️⃣ Importing the Module

Python provides a built-in module called **zipfile**.

```python
import zipfile
```

No installation is required.

---

# 5️⃣ Compression Types (Very Important)

Python supports different compression methods.

The two most commonly used are:

## A. ZIP_STORED

### Definition

Stores files without compression.

### Analogy

📚 Putting books in a bag without folding them.

Size remains almost the same.

### Syntax

```python
zipfile.ZipFile(
    "files.zip",
    "w",
    compression=zipfile.ZIP_STORED
)
```

### Features

❌ No compression

❌ File size not reduced

✅ Faster to create

---

## B. ZIP_DEFLATED

### Definition

Compresses files to reduce their size.

### Analogy

👕 Folding clothes before putting them into a bag.

Takes less space.

### Syntax

```python
zipfile.ZipFile(
    "files.zip",
    "w",
    compression=zipfile.ZIP_DEFLATED
)
```

### Features

✅ Reduces file size

✅ Best for sharing and storage

❌ Slightly slower than ZIP_STORED

---

# 6️⃣ Difference Between ZIP_STORED and ZIP_DEFLATED

| Feature | ZIP_STORED | ZIP_DEFLATED |
|----------|------------|--------------|
| Compression | ❌ No | ✅ Yes |
| File Size | Same | Smaller |
| Speed | Faster | Slightly Slower |
| Storage Saving | ❌ No | ✅ Yes |

---

# 7️⃣ Creating a ZIP File

## Syntax

```python
with zipfile.ZipFile(
    "files.zip",
    "w",
    compression=zipfile.ZIP_DEFLATED
) as z:

    z.write("file1.txt")
    z.write("file2.txt")
```

### Explanation

- `"files.zip"` → ZIP file name
- `"w"` → Write mode
- `z.write()` → Add files into ZIP

---

# 8️⃣ Example: Create a Compressed ZIP File

```python
import zipfile

with zipfile.ZipFile(
    "compressed.zip",
    "w",
    compression=zipfile.ZIP_DEFLATED
) as z:

    z.write("data.txt")
    z.write("students.csv")

print("Compressed ZIP created")
```

### Output

```text
Compressed ZIP created
```

---

# 9️⃣ Example: Create ZIP Without Compression

```python
import zipfile

with zipfile.ZipFile(
    "normal.zip",
    "w",
    compression=zipfile.ZIP_STORED
) as z:

    z.write("data.txt")

print("ZIP created without compression")
```

### Output

```text
ZIP created without compression
```

# 🔟 Extracting (Unzipping) Files

## Definition

Unzipping means extracting files from a ZIP archive.

### Analogy

🎒 Taking books out of a school bag.

---

## Extract All Files

### Syntax

```python
with zipfile.ZipFile("files.zip", "r") as z:
    z.extractall()
```

---

## Example

```python
import zipfile

with zipfile.ZipFile("compressed.zip", "r") as z:
    z.extractall()

print("Files extracted")
```

### Output

```text
Files extracted
```

---

# 1️⃣1️⃣ Extract a Specific File

### Syntax

```python
z.extract("file1.txt")
```

### Example

```python
import zipfile

with zipfile.ZipFile("compressed.zip", "r") as z:
    z.extract("data.txt")

print("File extracted")
```

---

# 1️⃣2️⃣ View Files Inside ZIP

Sometimes we want to see what files are present inside the ZIP.

### Syntax

```python
with zipfile.ZipFile("files.zip", "r") as z:
    print(z.namelist())
```

### Example

```python
import zipfile

with zipfile.ZipFile("compressed.zip", "r") as z:
    print(z.namelist())
```

### Output

```python
['data.txt', 'students.csv']
```

---

# 1️⃣3️⃣ Add More Files to Existing ZIP

Use append mode `"a"`.

### Syntax

```python
with zipfile.ZipFile("files.zip", "a") as z:
    z.write("newfile.txt")
```

---

## Example

```python
import zipfile

with zipfile.ZipFile("compressed.zip", "a") as z:
    z.write("newfile.txt")

print("File added")
```

---

# 1️⃣4️⃣ ZIP File Modes

| Mode | Meaning |
|--------|---------|
| `"w"` | Create new ZIP |
| `"r"` | Read existing ZIP |
| `"a"` | Add files to existing ZIP |

---

# 1️⃣5️⃣ Important Methods

| Method | Purpose |
|----------|----------|
| `write()` | Add file to ZIP |
| `extract()` | Extract one file |
| `extractall()` | Extract all files |
| `namelist()` | Show all file names |
| `close()` | Close ZIP file |

---

# 1️⃣6️⃣ Complete Example

```python
import zipfile

with zipfile.ZipFile(
    "backup.zip",
    "w",
    compression=zipfile.ZIP_DEFLATED
) as z:

    z.write("data.txt")
    z.write("students.csv")

print("ZIP created")

with zipfile.ZipFile("backup.zip", "r") as z:
    print("Files inside ZIP:")
    print(z.namelist())

    z.extractall()

print("ZIP extracted")
```

# 1️⃣8️⃣ Important Points to Remember

✅ `zipfile` is a built-in module.

✅ Files must exist before adding them to ZIP.

✅ Use `ZIP_DEFLATED` to reduce file size.

✅ Use `"a"` mode to add new files.

✅ Use `namelist()` to view files inside ZIP.

✅ A ZIP file can contain multiple files and folders.

✅ Always use `with` statement because it automatically closes the ZIP file.
