# Chapter 1: Introduction to `eval()`

## What is `eval()`?

### Definition

`eval()` is a **built-in Python function** that takes a string containing a valid Python expression, evaluates it as Python code, and returns the result.

---

### Simple Meaning

The `eval()` function reads a string, understands it as a Python expression, executes it, and returns the actual result.

---

### One-Line Definition

> **`eval()` evaluates a string as a Python expression and returns the result in its original data type.**

---

## Real-World Analogy

Imagine a teacher writes:

```
10 + 20
```

A student reads it, solves it, and answers:

```
30
```

Similarly,

`eval()` reads the string,

understands it as Python code,

and returns the calculated result.

---

# Chapter 2: Why Do We Need `eval()`?

Normally, Python's `input()` function returns **everything as a string**, even if the user enters a number.

---

## Example

```python
a = input("Enter a value: ")

print(a)
print(type(a))
```

### Input

```
10
```

### Output

```
10
<class 'str'>
```

---

## Observation

Even though the user entered a number,

Python stored it as:

```
"10"
```

which is a string.

---

# Chapter 3: Solution — Using `eval()`

Instead of:

```python
a = input()
```

write:

```python
a = eval(input())
```

---

## Example

```python
a = eval(input("Enter a value: "))

print(a)
print(type(a))
```

### Input

```
10
```

### Output

```
10
<class 'int'>
```

---

## Observation

`eval()` automatically converts the input into its correct Python data type.

---

# Chapter 4: How `eval()` Works

Suppose the user enters:

```
10 + 20
```

Python first receives:

```
"10 + 20"
```

Then,

`eval()` executes it as a Python expression.

Result:

```
30
```

---

## Flow Diagram

```
User Input

      │

      ▼

"10 + 20" (String)

      │

      ▼

eval()

      │

      ▼

30 (Integer)
```

# Chapter 6: Difference Between `input()` and `eval()`

| Function | Returns | Executes Expression? |
|-----------|----------|----------------------|
| `input()` | Always a string (`str`) | ❌ No |
| `eval()` | Original Python data type | ✅ Yes |

---

## Example

### Using `input()`

```python
a = input("Enter a: ")
b = input("Enter b: ")

print(a + b)
```

Input

```
10
20
```

Output

```
1020
```

---

### Why?

Because

```
"10" + "20"
```

means string concatenation.

---

## Using `eval()`

```python
a = eval(input("Enter a: "))
b = eval(input("Enter b: "))

print(a + b)
```

Input

```
10
20
```

Output

```
30
```

---

# Chapter 11: Disadvantages of `eval()`

`eval()` can execute **any valid Python code**.

This makes it **dangerous** if used with untrusted user input.

---

## Example

```python
eval("print('Hello')")
```

Output

```
Hello
```

---

This works because `eval()` executes the expression.

---

# Chapter 12: Security Warning (Very Important)

⚠ **Never use `eval()` with input from untrusted users in real-world applications.**

A malicious user could execute harmful Python code.

For example, they might try to access files, run commands, or perform unauthorized actions.

Because of this security risk, professional Python developers generally avoid using `eval()` for user input.

---

## Real-World Analogy

Imagine giving a stranger the keys to your house.

They could:

- Turn on the lights.
- Open every room.
- Move your belongings.

Similarly,

`eval()` gives the input the ability to execute Python code.

---

# Chapter 13: Safe Alternatives

Instead of `eval()`, use specific conversion functions.

---

## Integer

```python
num = int(input("Enter number: "))
```

---

## Float

```python
price = float(input("Enter price: "))
```

---

## String

```python
name = input("Enter name: ")
```

---

## Boolean (Example)

```python
value = input("Enter True or False: ")

flag = value == "True"
```

---

These methods are much safer because they convert only to the expected data type.

---

# Chapter 14: When Should You Use `eval()`?

## Use `eval()` When

- Learning Python.
- Understanding expressions.
- Accepting multiple values in one line.
- Solving classroom exercises.
- Writing small experimental programs.

---

## Avoid `eval()` When

- Building real-world applications.
- Accepting input from unknown users.
- Developing web applications.
- Writing banking software.
- Handling confidential or sensitive data.

# Interview Questions

### 1. What is `eval()`?

`eval()` is a built-in Python function that evaluates a string as a Python expression and returns the result in its original data type.

---

### 2. What is the difference between `input()` and `eval()`?

`input()` always returns a string, while `eval()` evaluates the input as a Python expression and returns its actual data type.

---

### 3. Why is `eval()` considered dangerous?

Because it executes Python code contained in the input. If the input comes from an untrusted source, it can lead to serious security vulnerabilities.

---

### 4. Can `eval()` evaluate mathematical expressions?

Yes.

Example

```python
eval("10+20")
```

returns

```
30
```

---

### 5. What data types can `eval()` return?

Depending on the input, it can return:

- `int`
- `float`
- `complex`
- `bool`
- `list`
- `tuple`
- `set`
- `dict`
- Other valid Python objects

---

### 6. What is the safest alternative to `eval()` for numeric input?

Use:

```python
int(input())
```

or

```python
float(input())
```

These functions convert only to the intended data type and do not execute arbitrary Python code.

---

# Real-World Teaching Line

Before using `eval()`, ask yourself:

- Do I need to evaluate a Python expression? → `eval()` may be appropriate for learning or controlled environments.
- Am I only reading a number? → Use `int()` or `float()`.
- Am I accepting input from unknown users? → **Avoid `eval()`** and use safe conversion functions instead.

Choosing the correct approach makes your Python programs safer and more reliable.

---

# Final One-Line Summary

> **`eval()` evaluates a string as a Python expression and returns the result in its original data type, but because it can execute arbitrary Python code, it should be used only in trusted environments and avoided for untrusted user input.**
